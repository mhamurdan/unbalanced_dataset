```python
def grudge_check(click):
    tablexxx= pd.DataFrame([])
    names=click.select_dtypes(include=['O']).columns
    for c in names:
        tablex2=pd.crosstab(index=click[c],columns='perc')
        tablex2['perc']=tablex2['perc']/tablex2['perc'].sum()
        tablexxx=tablexxx.append(tablex2)
    return tablexxx.loc[(tablexxx['perc'] < 0.05)]
```

```python
grudge_check(cat_train)
```

```python
## convert categorical variables lt 5 perc to others

def grudge(click):
    names=click.select_dtypes(include=['O']).columns
    for c in names:
        series = pd.value_counts(click[c])
        mask = (series/series.sum() * 100).lt(5)
        click[c] = np.where(click[c].isin(series[mask].index),'Other',click[c])

## convert categorical variables selected in train in test aswell
##        
def grudge_for_test(train_data, test_data):
    names=train_data.select_dtypes(include=['O']).columns
    for c in names:
        series = pd.value_counts(train_data[c])
        mask = series.gt(0)
        test_data[c] = np.where(np.logical_not(test_data[c].isin(series[mask].index)),'Other',test_data[c])


grudge(cat_train)

grudge_for_test(cat_train,cat_test)
```
